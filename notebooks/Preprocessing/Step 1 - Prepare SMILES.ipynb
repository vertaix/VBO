```python
import os
import pandas as pd

import math
```

```python
path = "../data/DB_10042.csv"

df = pd.read_csv(path, index_col=0)
df
```

```python
list(df.columns)
```

```python
num_node_smiles = []
node_smiles = []

num_linker_smiles = []
linker_smiles = []

for row_i, row in df.iterrows():
    # node
    early_break_flag = False
    for node_i in range(1, 6):
        this_node_smiles = row[f"node{node_i}"]
        if isinstance(this_node_smiles, str):
            node_smiles.append(this_node_smiles)
        else:
            for cont_node_i in range(node_i + 1, 6):
                assert math.isnan(row[f"node{cont_node_i}"])
            num_node_smiles.append(node_i - 1)
            early_break_flag = True
            break
    if not early_break_flag:
        num_node_smiles.append(5)
    
    # linker
    early_break_flag = False
    for linker_i in range(1, 21):
        this_linker_smiles = row[f"linker{linker_i}"]
        if isinstance(this_linker_smiles, str):
            linker_smiles.append(this_linker_smiles)
        else:
            for cont_linker_i in range(linker_i + 1, 21):
                assert math.isnan(row[f"linker{cont_linker_i}"])
            num_linker_smiles.append(linker_i - 1)
            early_break_flag = True
            break
    if not early_break_flag:
        num_linker_smiles.append(20)
            
    assert sum(num_node_smiles) == len(node_smiles)
    assert sum(num_linker_smiles) == len(linker_smiles)
```

```python
len(num_node_smiles), len(node_smiles), len(num_linker_smiles), len(linker_smiles)
```

```python
sum(num_node_smiles), len(node_smiles), sum(num_linker_smiles), len(linker_smiles)
```

```python
assert sum(num_node_smiles) == len(node_smiles)
assert sum(num_linker_smiles) == len(linker_smiles)
```

```python
df["num_node_smiles"] = num_node_smiles
df["num_linker_smiles"] = num_linker_smiles
```

```python
df.to_csv(path.split(".csv")[0] + "_w_num_smiles.csv")
```

```python
with open(f"../data/{path.split('.csv')[0]}_node_smiles.txt", "w") as f:
    for smiles in node_smiles:
        f.write(smiles + "\n")
```

```python
with open(f"../data/{path.split('.csv')[0]}_linker_smiles.txt", "w") as f:
    for smiles in linker_smiles:
        f.write(smiles + "\n")
```
